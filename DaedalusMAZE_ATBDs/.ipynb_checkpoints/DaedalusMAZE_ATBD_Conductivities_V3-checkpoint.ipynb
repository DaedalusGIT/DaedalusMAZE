#### Introduction
<a class="anchor" id="Introduction"></a>
<p style='text-align: justify;'>
Pedersen conductivity ($σ_P$) is required in Joule heating calculations, but is also extremely important in ionospheric and magnetospheric modelling. Pedersen conductivity cannot be obtained by direct measurement, as it depends on a large number of terms, such as composition of different ion species (in number densities $N_i$) and the various collision rates between ion species and neutral species. These, in turn, depend on the ion and electron temperatures ($T_i$ and $T_n$), the neutral densities of the neutral species ($N_n$), and the ion species. In calculations of the Pedersen conductivity, the ratio ($r_i$) of each species’ collision rate versus its gyrofrequency needs to be estimated. The collision rate will depend on the species’ density, thus neutral composition needs to be measured independently through mass spectrometry.

#### Conductivity estimation methodology
<a class="anchor" id="Conductivity-estimation-methodology"></a>
<p style='text-align: justify;'>
The Pedersen conductivity, $σ_P$, Hall conductivity, $σ_H$, and Parallel conductivity,  $σ_{\parallel}$ can be calculated as:

$$  \sigma_{P} = \frac{q_{e}}{B} [ Ν_{O_2^+}\frac{\nu_{O_{2}^{+}}Ω_{O_{2}^{+}}}{\nu_{O_{2}^{+}}^{2}+Ω_{O_{2}^{+}}^{2} } +Ν_{O^+}\frac{\nu_{O^{+}}Ω_{O^{+}}}{\nu_{O^{+}}^{2}+Ω_{O^{+}}^{2}} +Ν_{NO^+}\frac{\nu_{NO^{+}}Ω_{NO^{+}}}{\nu_{NO^{+}}^{2}+Ω_{NO^{+}}^{2} }+
										   Ν_{e}\frac{\nu_{en\perp}Ω_{e}}{\nu_{en\perp}^{2}+Ω_{e}^{2} } 
										 ] \label{eq:Pedersen} \tag{1}$$
$$  \sigma_{H} = \frac{Ν_{e}q_{e}}{B} [ -\frac{\nu_{in}Ω_{i}}{\nu_{in}^{2}+Ω_{i}^{2} } +
										   \frac{\nu_{en\perp}Ω_{i}}{\nu_{en\perp}^{2}+Ω_{e}^{2} } 
										 ] \label{eq:Hall} \tag{2}$$
$$  \sigma_{\parallel} = \frac{Ν_{e}\,q_{e}^2}{m_e\,(\nu_{en\parallel}+\nu_{ei\parallel})} \label{eq:Parallel} \tag{3}$$


Where $$\Omega_{j}=\frac{eB}{m_j}\label{eq:Omega} \tag{4}$$ is the gyrofrequency with the subscript $j$ denoting electrons ($j=e$) or ions ($j=i$), $q_{e}$ is the elementary electric charge ($1.602176634 \times 10^{-19} C $) and $N_{j}$ denotes the electron ($j=e$) or ion ($j=i$) density (in $m^{-3}$),and $\nu$ is the collision frequency.
</p>

<p style='text-align: justify;'>
The collision rates depend on a number of terms, such as: on the density and composition of the ion and neutral species, which need to be measured independently through mass spectrometry; on the ion and electron temperatures; and on the values for collision cross sections. The latter are calculated primarily through laboratory experiments of ion-neutral collisions. However, these may have systematic uncertainties in the upper atmosphere, and their accuracy has never been evaluated in-situ. These measurements, through the comparison of the two methodologies described above, will also allow a new estimate of collisional cross sections, which are needed and widely used in simulations, analysis of radar data etc. An overview of the various parameters used in conductivity calculations is given in the following tables:

<p style='text-align: justify;'>
So, if we want to evaluate every parameter that goes into a Joule heating calculation in a local volume of space, the geophysical observables required are total density, the neutral winds, the ion drifts, the ion density, ion temperature, electron temperature, ion composition, the total neutral composition which is a separate measurement that can be used as a check, the magnetic field and the DC electric fields.
In the system of the neutral atmosphere, conductivities will be estimated by Eq. (1) to (3) listed above.


The error of both quantities can be computed by error propagation formula, applied first to Eq. (5), to derive the error in conductivity, and then to Eq. (4), to derive the error in Joule heating. The error in the higher level, L3, products depend on the individual errors of the observed quantities, L2 products (δE, δvn, δB, the constituents of δσ), to be provided by instrument teams (WP2.1, WP3.4) in relation to state of the art resolution and accuracy (WP1.1), consolidated product requirements (WP1.4), and performance metrics (WP3.4).

#### Error Analysis for Conductivity estimates
<a class="anchor" id="Error-Analysis_Code"></a>

    Error Analysis for Pedersen Conductivity
    
$$  \sigma_{P} = \frac{q_{e}}{B} [ Ν_{O_2^+}\frac{\nu_{O_{2}^{+}}Ω_{O_{2}^{+}}}{\nu_{O_{2}^{+}}^{2}+Ω_{O_{2}^{+}}^{2} } +Ν_{O^+}\frac{\nu_{O^{+}}Ω_{O^{+}}}{\nu_{O^{+}}^{2}+Ω_{O^{+}}^{2}} +Ν_{NO^+}\frac{\nu_{NO^{+}}Ω_{NO^{+}}}{\nu_{NO^{+}}^{2}+Ω_{NO^{+}}^{2} }+
										   Ν_{e}\frac{\nu_{en\perp}Ω_{e}}{\nu_{en\perp}^{2}+Ω_{e}^{2} } 
										 ] \label{eq:Pedersen} \tag{1}$$

* $$ d\sigma_{P}^2=\mid\frac{\partial\sigma_{P}}{\partial B}\mid^2 dB^2+\mid\frac{\partial\sigma_{P}}{\partial N_e}\mid^2 dN_{e}^{2}+\mid\frac{\partial\sigma_{P}}{\partial N_{O^+}}\mid^2 dN_{O^+}+ \mid\frac{\partial\sigma_{P}}{\partial N_{O_2^+}}\mid^2 d N_{O_2^+}+ \mid\frac{\partial\sigma_{P}}{\partial N_{NO^+}}\mid^2 d N_{NO^+} +\mid\frac{\partial\sigma_{P}}{\partial\nu_{O_{2}^{+}}}\mid^2 d\nu_{O_{2}^{+}}^2+\\
\mid\frac{\partial\sigma_{P}}{\partial\nu_{O^{+}}}\mid^2 d\nu_{O^{+}}^2+\mid\frac{\partial\sigma_{P}}{\partial\nu_{NO^{+}}}\mid^2 d\nu_{NO^{+}}^2+\mid\frac{\partial\sigma_{P}}{\partial\nu_{en\perp}}\mid^2 d\nu_{en\perp}^2+\mid\frac{\partial\sigma_{P}}{\partial\Omega_{O{2}^{+}}}\mid^2 d\Omega_{O{2}^{+}}^2+\mid\frac{\partial\sigma_{P}}{\partial\Omega_{NO^{+}}}\mid^2 d\Omega_{NO^{+}}^2+\mid\frac{\partial\sigma_{P}}{\partial\Omega_{en\perp}}\mid^2 d\Omega_{en\perp}^2\tag{29} $$



* $$\mid\frac{\partial\sigma_{P}}{\partial B}\mid^2=(-q_e \frac{1}{B^2} (\frac{N_{NO^+} \Omega_{NO^+} \nu_{NO^+}}{\Omega_{NO^+}^2 + \nu_{NO^+}^2} + \frac{N_{O_2^+}\Omega_{O_2^+} \nu_{O_2^2}}{\Omega_{O_2^+}^2 + \nu_{O_2^+}^2} +\frac{N_{O^+} \Omega_{O^+}\nu_{O^+}}{(\Omega_{O^+}^2 + V_{O^+}^2)} + \frac{\Omega_e \nu_{e\perp}}{(\Omega_e^2 + \nu_{e\perp}^2)})^2 \tag{30}$$


* $$\mid\frac{\partial\sigma_{P}}{\partial N_e}\mid^2= [\frac{\Omega_{e} \nu_{e \perp} q_{e}}{B \left(\Omega_{e}^{2} + \nu_{e \perp}^{2}\right)}]^2\tag{31}$$


* $$\mid\frac{\partial\sigma_{P}}{\partial N_{O^+}}\mid^2= [\frac{\Omega_{O^+} \nu_{O^+} q_{e}}{B \left(\Omega_{O^+}^{2} + \nu_{O^+}^{2}\right)}]^2\tag{32}$$


* $$\mid\frac{\partial\sigma_{P}}{\partial N_{O_2^+}}\mid^2= [\frac{\Omega_{O_2^p} \nu_{O_2^+} q_{e}}{B \left(\Omega_{O_2^+}^{2} + \nu_{O_2^+}^{2}\right)}]^2\tag{33}$$
 
 
* $$\mid\frac{\partial\sigma_{P}}{\partial N_{NO^+}}\mid^2= [\frac{\Omega_{NO^+} \nu_{NO^+} q_{e}}{B \left(\Omega_{NO^+}^{2} + \nu_{NO^+}^{2}\right)}]^2\tag{34}$$

* $$\mid\frac{\partial\sigma_{P}}{\partial\nu_{en\perp}}\mid^2= [\frac{q_{e} \left(- \frac{2 N_e \Omega_{e} \nu_{e\perp}^{2}}{\left(\Omega_{e}^{2} + \nu_{e \perp}^{2}\right)^{2}} + \frac{N_e \Omega_{e}}{\Omega_{e}^{2} + \nu_{e\perp}^{2}}\right)}{B}]^2\tag{35}$$

* $$\mid\frac{\partial\sigma_{P}}{\partial\nu_{O^{+}}}\mid^2= [\frac{q_{e} \left(- \frac{2 N_{O^+} \Omega_{O^+} \nu_{O^+}^{2}}{\left(\Omega_{O^+}^{2} + \nu_{O^+}^{2}\right)^{2}} + \frac{N_{O^+} \Omega_{O^+}}{\Omega_{O^+}^{2} + \nu_{O^+}^{2}}\right)}{B}]^2\tag{36}$$

* $$\mid\frac{\partial\sigma_{P}}{\partial\nu_{O_{2}^{+}}}\mid^2= [\frac{q_{e} \left(- \frac{2 N_{O_2^+} \Omega_{O_2^+} \nu_{O_2^+}^{2}}{\left(\Omega_{O_2^+}^{2} + \nu_{O_2^+}^{2}\right)^{2}} + \frac{N_{O_2^+} \Omega_{O2p}}{\Omega_{O_2^+}^{2} + \nu_{O_2^+}^{2}}\right)}{B}]^2\tag{37}$$

* $$\mid\frac{\partial\sigma_{P}}{\partial\nu_{NO^{+}}}\mid^2= [\frac{q_{e} \left(- \frac{2 NO_{+} \Omega_{NO^+} V_{NO^+}^{2}}{\left(\Omega_{NO^+}^{2} + V_{NO^+}^{2}\right)^{2}} + \frac{NO_{+} \Omega_{NOp}}{\Omega_{NO^+}^{2} + V_{NO^+}^{2}}\right)}{B}]^2\tag{38}$$

* $$\mid\frac{\partial\sigma_{P}}{\partial\Omega_{en\perp}}\mid^2= [\frac{q_{e} \left(- \frac{2 N_e \Omega_{en\perp}^{2} \nu_{e \perp}}{\left(\Omega_{e}^{2} + \nu_{e \perp}^{2}\right)^{2}} + \frac{N_e \nu_{e \perp}}{\Omega_{e}^{2} + \nu_{e \perp}^{2}}\right)}{B}]^2 \tag{39}$$

* $$\mid\frac{\partial\sigma_{P}}{\partial\Omega_{O^{+}}}\mid^2= [\frac{q_{e} \left(- \frac{2 N_{O^+} \Omega_{O^+}^{2} \nu_{O^+}}{\left(\Omega_{O^+}^{2} + \nu_{O^+}^{2}\right)^{2}} + \frac{N_{O^+} \nu_{O^+}}{\Omega_{O^+}^{2} + \nu_{O^+}^{2}}\right)}{B}]^2\tag{40}$$

* $$\mid\frac{\partial\sigma_{P}}{\partial\Omega_{O_{2}^{+}}}\mid^2= [\frac{q_{e} \left(- \frac{2 N_{O_2^+} \Omega_{O_2^+}^{2} \nu_{O_2^+}}{\left(\Omega_{O_2^+}^{2} + \nu_{O_2^+}^{2}\right)^{2}} + \frac{N_{O_2^+} \nu_{O_2^+}}{\Omega_{O_2^+}^{2} + \nu_{O_2^+}^{2}}\right)}{B}]^2\tag{41}$$

* $$\mid\frac{\partial\sigma_{P}}{\partial\Omega_{NO^{+}}}\mid^2= [\frac{q_{e} \left(- \frac{2 N_{NO^+} \Omega_{NO_+}^{2} \nu_{NO_+}}{\left(\Omega_{NO_+}^{2} + \nu_{NO_+}^{2}\right)^{2}} + \frac{N_{NO^+} \nu_{NO_+}}{\Omega_{NO_+}^{2} + \nu_{NO_+}^{2}}\right)}{B}]^2  \tag{42}$$


* $$ d\Omega_{e}^2=\mid\frac{\partial\Omega_{e}}{\partial B}\mid^2 dB^2=(\frac{q_e}{m_{e}} dB)^2 \tag{43}$$

* $$ d\Omega_{O^{+}}^2=\mid\frac{\partial\Omega_{O^{+}}}{\partial B}\mid^2 dB^2= (\frac{q_e}{m_{O^+}} dB)^2 \tag{44}$$

* $$ d\Omega_{O_{2}^{+}}^2=\mid\frac{\partial\Omega_{O{2}^{+}}}{\partial B}\mid^2 dB^2=( \frac{q_e}{m_{O_{2}^{+}}}  dB)^2\tag{45}$$

* $$ d\Omega_{NO^{+}}^2=\mid\frac{\partial\Omega_{NO^{+}}}{\partial B}\mid^2 dB^2=(\frac{q_e}{m_{NO^+}} dB)^2\tag{46}$$



* $$ d\nu_{O_{2}^{+}}^2= {d\nu_{O_{2}^{+}-O}}^2+{d\nu_{O_{2}^{+}-O_2}}^2+{d\nu_{O_{2}^{+}-N_2}}^2 \tag{47}$$


$${d\nu_{O_{2}^{+}-O}}^2=\mid\frac{\partial\nu_{O_{2}^{+}-O}}{\partial N_O}\mid^2 d N_O^2 + \mid\frac{\partial\nu_{O_{2}^{+}-O}}{\partial T_n}\mid^2 d T_n^2 + + \mid\frac{\partial\nu_{O_{2}^{+}-O}}{\partial T_i}\mid^2 d T_i^2\tag{48}$$

$$ \mid\frac{\partial\nu_{O_{2}^{+}-O}}{\partial N_O}\mid^2 = (\frac{1.8 \cdot 10^{-16}}{\left(\frac{T_{i}}{1000} + \frac{T_{n}}{1000}\right)^{0.19}})^2 \tag{49}$$

$$ \mid\frac{\partial\nu_{O_{2}^{+}-O}}{\partial T_n}\mid^2 = (- \frac{3.42 \cdot 10^{-20} N_{O}}{\left(\frac{T_{i}}{1000} + \frac{T_{n}}{1000}\right)^{1.19}})^2 \tag{50}$$

$$ \mid\frac{\partial\nu_{O_{2}^{+}-O}}{\partial T_i}\mid^2 = (- \frac{3.42 \cdot 10^{-20} N_{O}}{\left(\frac{T_{i}}{1000} + \frac{T_{n}}{1000}\right)^{1.19}})^2 \tag{51}$$


$${d\nu_{O_{2}^{+}-O_2}}^2=\mid\frac{\partial\nu_{O_{2}^{+}-O_2}}{\partial N_{O_2}}\mid^2 d N_{O_2}^2\tag{52}$$

$$\mid\frac{\partial\nu_{O_{2}^{+}-O_2}}{\partial N_{O_2}}\mid^2=(5.2 \cdot 10^{-16})^2 \tag{53}$$

$${d\nu_{O_{2}^{+}-N_2}}^2=\mid\frac{\partial\nu_{O_{2}^{+}-N_2}}{\partial N_{N_2}}\mid^2 d N_{N_2}^2\tag{54}$$

$$\mid\frac{\partial\nu_{O_{2}^{+}-N_2}}{\partial N_{N_2}}\mid^2 =(4.3 \cdot 10^{-16})^2 \tag{55}$$

* $$ d\nu_{O^{+}}^2= {d\nu_{O^{+}-O}}^2+{d\nu_{O^{+}-O_2}}^2+{d\nu_{O^{+}-N_2}}^2 \tag{56}$$


* $${d\nu_{O^{+}-O}}^2=\mid\frac{\partial\nu_{O^{+}-O}}{\partial N_O}\mid^2 d N_O^2 + \mid\frac{\partial\nu_{O^{+}-O}}{\partial T_n}\mid^2 d T_n^2 + + \mid\frac{\partial\nu_{O^{+}-O}}{\partial T_i}\mid^2 d T_i^2 \tag{57}$$

$$\mid\frac{\partial\nu_{O^{+}-O}}{\partial N_O}\mid^2=(6.7 \cdot 10^{-16} \left(\frac{T_{i}}{1000} + \frac{T_{n}}{1000}\right)^{0.05} \left(- \frac{\log{\left(\frac{T_{i}}{1000} + \frac{T_{n}}{1000} \right)}}{\log{\left(10 \right)}} + 0.96\right)^{2})^2 \tag{58}$$
